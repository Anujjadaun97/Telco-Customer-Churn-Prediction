# Customer Churn Prediction

## Business Understanding

### Project Overview

Customer churn is one of the biggest challenges for subscription-based businesses. Acquiring a new customer is often more expensive than retaining an existing one. Therefore, identifying customers who are likely to leave allows companies to take proactive actions to improve retention.

This project develops a Machine Learning model that predicts whether a customer is likely to churn based on customer demographics, account information, subscribed services, and billing details.

The project follows the complete Data Science lifecycle:

- Business Understanding
- Data Cleaning
- Exploratory Data Analysis
- Feature Engineering
- Model Building
- Model Evaluation
- Business Recommendations

# Business Problem

Telecommunication companies experience customer churn every month, resulting in revenue loss and increased customer acquisition costs.

The company wants to answer the following business question:

> Which customers are most likely to churn, and what factors influence their decision to leave?

The objective is to build a predictive model that helps identify high-risk customers before they leave.

# Business Objectives

The objectives of this project are:

- Predict whether a customer will churn.
- Identify the major factors influencing churn.
- Help the business reduce customer attrition.
- Improve customer retention strategies.
- Increase Customer Lifetime Value (CLV).
- Reduce Customer Acquisition Cost (CAC).

# Stakeholders

This project provides value to multiple business teams.

| Stakeholder | Business Use |
|-------------|--------------|
| Executive Leadership | Monitor churn trends and business performance |
| Marketing Team | Launch targeted retention campaigns |
| Customer Success Team | Contact high-risk customers proactively |

| Sales Team | Promote long-term contracts and upgrades |

# Success Criteria

The project will be considered successful if:

- The model accurately identifies customers at risk of churning.
- The business gains actionable insights into churn drivers.
- High-risk customers can be identified before they leave.
- Business recommendations can be implemented to improve retention.

# Business Hypotheses

Before analyzing the data, the following hypotheses are proposed:

1. Customers on month-to-month contracts are more likely to churn.
2. Customers with shorter tenure are more likely to churn.
3. Customers paying higher monthly charges are more likely to churn.
4. Fiber Optic customers have higher churn.
5. Customers using Electronic Check have higher churn.
6. Customers without partners or dependents are more likely to churn.
7. Customers with premium support services are less likely to churn.

# Machine Learning Problem

Problem Type:

Binary Classification

Target Variable:

Churn

Target Classes:

- Yes
- No

Evaluation Metrics:

- Precision
- Recall
- F1 Score
- ROC-AUC

Since churn prediction is an imbalanced classification problem, Recall and ROC-AUC are more important than Accuracy alone.
