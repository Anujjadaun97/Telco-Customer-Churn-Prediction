### Churn Distribution

```python
fig, axes = plt.subplots(1,2, figsize=(12,5))

#count plot
df['Churn'].value_counts().plot.bar(ax=axes[0], color=['steelblue', 'tomato'], edgecolor='black')
axes[0].set_title('Churn Count')
axes[0].set_xlabel('Churn')
axes[0].set_ylabel('Count')

# Pie chart
df['Churn'].value_counts().plot.pie(ax=axes[1], autopct='%1.1f%%', colors=['steelblue', 'tomato'], startangle=90)
axes[1].set_title('Churn Proportion')
axes[1].set_ylabel('')
plt.tight_layout()
plt.show()                                   
```

### Categorical Columns VS Churn

```python
cat_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
            'PhoneService', 'InternetService', 'Contract',
            'PaymentMethod', 'PaperlessBilling']

fig, axes = plt.subplots(3, 3, figsize=(18, 18))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(data=df, x=col, hue='Churn', palette=['steelblue', 'tomato'], ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=15)

plt.suptitle('Categorical Features vs Churn', fontsize=16, y=1.01)
plt.tight_layout()
plt.show()
```

###  Numeric Columns Distribution (Histogram + KDE)

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for i, col in enumerate(num_cols):
    sns.histplot(data=df, x=col, hue='Churn', kde=True,
                 palette=['steelblue', 'tomato'], ax=axes[i], bins=30)
    axes[i].set_title(f'{col} Distribution by Churn')

plt.tight_layout()
plt.show()
```

###  Boxplots — Numeric vs Churn

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x='Churn', y=col,
                palette=['steelblue', 'tomato'], ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')

plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python
# Encode for correlation
df_encoded = df.copy()
df_encoded['Churn'] = df_encoded['Churn'].map({'Yes': 1, 'No': 0})

# Select numeric + encoded target
corr = df_encoded.select_dtypes(include='number').corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

### Churn Rate by Contract Type (Business Insight)

```python
churn_rate = df.groupby('Contract')['Churn'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
churn_rate.columns = ['Contract', 'Churn Rate (%)']

plt.figure(figsize=(7, 4))
sns.barplot(data=churn_rate, x='Contract', y='Churn Rate (%)',
            palette='Reds_d', edgecolor='black')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate (%)')
plt.tight_layout()
plt.show()
```
